```
# Shared Input Layer
from keras.utils import plot_model
from keras.models import Model
from keras.layers import Input
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.convolutional import Conv2D
from keras.layers.pooling import MaxPooling2D
from keras.layers.merge import concatenate
# input layer
visible = Input(shape=(128,128,1))
#fs
conv1 = Conv2D(32, kernel_size=(1,1), activation='relu')(visible)
#t1
conv8 = Conv2D(32, kernel_size=(1,1), activation='relu')(conv1)
flat1 = Flatten()(conv8)

#t2
conv2 = Conv2D(32, kernel_size=(1,1), activation='relu')(conv1)
conv5 = Conv2D(32, kernel_size=(3,3), activation='relu')(conv2)
flat2 = Flatten()(conv5)

#t3
conv3 = Conv2D(32, kernel_size=(1,1), activation='relu')(conv1)
conv6 = Conv2D(32, kernel_size=(5,5), activation='relu')(conv3)
flat3 = Flatten()(conv6)

#t4
conv4 = Conv2D(32, kernel_size=(3,3), activation='relu')(conv1)
conv7 = Conv2D(32, kernel_size=(1,1), activation='relu')(conv4)
flat4 = Flatten()(conv7)

# merge feature extractors
merge = concatenate([flat1, flat2, flat3, flat4])

# interpretation model
output = Dense(1, activation='sigmoid')(merge)
model = Model(inputs=[visible], outputs=output)
# summarize layers
print(model.summary())

# plot graph
plot_model(model, to_file='1.png')
```

```
# Shared Input Layer
from keras.utils import plot_model
from keras.models import Model
from keras.layers import Input
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.convolutional import Conv2D
from keras.layers.pooling import MaxPooling2D
from keras.layers.merge import concatenate
# input layer
visible = Input(shape=(227,227,3))
#fs
conv1 = Conv2D(96, (11,11),strides=4, padding='same')(visible)
pool1 = MaxPooling2D(pool_size=(3, 3), strides=2)(conv1)
conv2 = Conv2D(256, (5,5),  activation='relu')(pool1)
pool2 = MaxPooling2D(pool_size=(3, 3), strides=2)(conv2)
conv3 = Conv2D(384, (3,3), activation='relu')(pool2)
conv4 = Conv2D(384, (3,3), activation='relu')(conv3)
conv5 = Conv2D(356, (3,3), activation='relu')(conv4)
pool3 = MaxPooling2D(pool_size=(3, 3), strides=2)(conv5)

flat1 = Flatten()(pool3)
#flat2 = Flatten()(flat1)
# merge feature extractors
#merge = concatenate([flat1, flat2])
# interpretation model
hidden1 = Dense(10, activation='relu')(flat1)
hidden2 = Dense(10, activation='relu')(hidden1)
output = Dense(1000, activation='softmax')(hidden2)
model = Model(inputs=[visible], outputs=output)
# summarize layers
print(model.summary())

# plot graph
plot_model(model, to_file='2.png')
```
