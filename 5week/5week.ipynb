```python
# Color 이미지 >> 흑백 이미지
from numpy import expand_dims
from numpy import asarray
from PIL import Image

# load the image
img = Image.open('hub.jpg')
# convert the image to grayscale
img_gray = img.convert(mode='L')

# convert to numpy array
data_gray = asarray(img_gray)
data_color = asarray(img)

print(data_gray.shape)
print(data_color.shape)
```

```python
# add channels first
data_first = expand_dims(data_gray, axis=0)
print(data_first.shape)
# add channels last
data_last = expand_dims(data_gray, axis=2)
print(data_last.shape)
```

```python
from keras import backend
if backend.image_data_format() == 'channels_last':
  backend.set_image_data_format('channels_first')
  print(backend.image_data_format())
else:
  backend.set_image_data_format('channels_last')
  print(backend.image_data_format())
```
